Some definitions: <br>
- The gain for each period is the end balance minus the start balance.

````python
data = ({"Initial": 1000}, {"Jan 2018":1200},{"Feb 2018":1400},{"Mar 2018":700},{"Apr 2018":800},{"May 2018":500})
data
````

````python
# 3) Your code here
import pandas as pd
import matplotlib.pyplot as plt

def bitcoin_invest(monthly_data):
    # Extracting keys and values 
    keys = [list(elem.keys())[0] for elem in monthly_data]
    balances = [list(elem.values())[0] for elem in monthly_data]

    # Creating a DataFrame with Date and Balance column
    df = pd.DataFrame({"Date": keys, "Balance": balances})

    # Creating the 'Return' column with percentage change
    df['Return'] = df['Balance'].pct_change()
    
    print(df)

    # Plotting the balance vs. month
    plt.figure(figsize=(10, 8))
    
    # Top plot
    ax1 = plt.subplot(2, 1, 1)
    ax1.plot(df['Date'][1:], df['Balance'][1:], marker='o', linestyle='-', color='blue') #starting with the ending balance of first month
    ax1.set_title('Balance vs. Month')
    ax1.set_xlabel('Month')
    ax1.set_ylabel('Balance')
    
    #Plotting Return vs. month
    # Bottom plot
    ax2 = plt.subplot(2, 1, 2, sharex=ax1)  # Specifying sharex=ax1 to share the x-axis
    bars = ax2.bar(df['Date'][1:], df['Return'][1:], color=df['Return'][1:].apply(lambda x: 'green' if x > 0 else 'red'))
    ax2.axhline(y=0, color='black', linestyle='--', linewidth=1)
    ax2.set_title('Return vs. Month')
    ax2.set_xlabel('Month')
    ax2.set_ylabel('Return')

    # Adding labels to bars
    for bar in bars:
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width() / 2, height,
                 f'{height:.2f}', ha='center', va='bottom' if height > 0 else 'top')

    plt.tight_layout()
    plt.show()

# Testing the function with provided data
data = ({"Initial": 1000}, {"Jan 2018":1200},{"Feb 2018":1400},{"Mar 2018":700},{"Apr 2018":800},{"May 2018":500})
bitcoin_invest(data)


    
````

### VCF file description (Summarized from version 4.1)

```
* The VCF specification:

VCF is a text file format which contains meta-information lines, a header line, and then data lines each containing information about a position in the genome. The format also can contain genotype information on samples for each position.

* Fixed fields:

There are 8 fixed fields per record. All data lines are **tab-delimited**. In all cases, missing values are specified with a dot (‘.’). 

1. CHROM - chromosome number
2. POS - position DNA nuceleotide count (bases) along the chromosome
3. ID - The unique identifier for each mutation
4. REF - reference base(s)
5. ALT - alternate base(s)
6. FILTER - filter status
7. QUAL - quality
8. INFO - a semicolon-separated series of keys with values in the format: <key>=<data>

```
### Applicable INFO field specifications

```
GENEINFO = <Gene name>
CLNSIG =  <Clinical significance>
CLNDN = <Disease name>
```

````python
# 4) Your code here - can use as many code blocks as you would like
import pandas as pd
from io import StringIO

def read_vcf(path):
    with open(path, 'r') as f:
        lines = [l for l in f if not l.startswith('##')]
    return pd.read_csv(
        StringIO(''.join(lines)),
        dtype={'CHROM': str, 'POS': int, 'ID': str, 'REF': str, 'ALT': str,
               'QUAL': str, 'FILTER': str, 'INFO': str},
        sep='\t',
        comment='#',  # Treat '#' as a comment character
        header=None
    ).rename(columns={0: 'CHROM'})

#path to clinvar_final.txt file
clinvar_file_path = 'C:\\Users\\tgeod\\Downloads\\UC Berkeley\\200- Python for Data Science\\clinvar_final.txt'

#read_vcf function to read the clinvar_final.txt file
df_clinvar = read_vcf(clinvar_file_path)

df_clinvar
````
